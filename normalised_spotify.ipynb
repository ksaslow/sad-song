### BRAZIL

```python
df = pd.DataFrame(data)
grouped_bra =df[df['country']== '-za-'].groupby(['date'])['date','country','Streams','danceability','energy','valence','tempo'].mean()
grouped_bra['HappyIndex'] = (grouped_bra.danceability*grouped_bra.energy*grouped_bra.valence)/grouped_bra.Streams
grouped_bra

#grouped2 = grouped[(grouped ['date'] >'2019-12-15')]
```

```python
brazil_index = grouped_bra.iloc[12:18, 5:6]

brazil_index.dtypes
```

```python
months_brazil= {'CountryName': ['Brazil','Brazil','Brazil','Brazil','Brazil','Brazil','Brazil','Brazil'],
        'Month': ['February', 'March', 'April', 'May', 'June', 'July', 'August', 'September']}
brazil = pd.DataFrame(months_brazil, columns=['CountryName', 'Month'])
brazil['HappyIndex'] = brazil_index
brazil
```
